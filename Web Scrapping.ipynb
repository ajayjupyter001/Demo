```python
cities=[]
preprocessed_cities=[]

def get_cities_and_preprocessing():
    country_html=open("C:/Users/BANAR/Desktop/DataScienceProjects/CarpricePrediction/Data/Country_data/country_name.html",encoding="utf-8")
    content=country_html.read()
    soup=BeautifulSoup(content,"html.parser")
    
    def get_unpreprocessed_cities():
        for j in soup.find_all("tr"):
            try:
                try:
                    city_name=j.td.next_sibling.next_sibling.b.a.text
                    cities.append(city_name)
                except:
                    city_name=j.td.next_sibling.next_sibling.a.text
                    cities.append(city_name)
            except:
                pass
    def done_preprocessing():
        for j in cities:
            lower_name=j.lower()
            split=lower_name.split(" ")
            if(len(split)>1):
                lower_name="-".join(split)
            preprocessed_cities.append(lower_name)
            
    get_unpreprocessed_cities()
    done_preprocessing()

get_cities_and_preprocessing()
```
